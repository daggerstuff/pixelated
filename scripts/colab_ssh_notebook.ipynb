```python
# =============================================================================
# CELL 1: KEEP ALIVE - Run this first, then minimize tab
# =============================================================================

import IPython
from google.colab import output
import time
import threading

# JavaScript keep-alive (clicks connect button periodically)
display(IPython.display.Javascript('''
  function KeepAlive() {
    console.log("Session keep-alive ping");
    var btn = document.querySelector("colab-connect-button");
    if (btn) {
      btn = btn.shadowRoot.querySelector("#connect");
      if (btn) btn.click();
    }
  }
  setInterval(KeepAlive, 55000);
'''))

# Python keep-alive (prints to keep kernel active)
def keep_alive():
    while True:
        time.sleep(60)
        print(".", end="", flush=True)

thread = threading.Thread(target=keep_alive, daemon=True)
thread.start()

print("✅ Keep-alive running. You can now minimize this tab.")
```
